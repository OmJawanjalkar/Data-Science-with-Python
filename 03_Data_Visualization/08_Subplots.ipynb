# Subplots

## Introduction

Subplots allow us to display multiple plots in a single figure.

Instead of creating separate windows for each chart, we can place multiple charts together for comparison and analysis.

Subplots are extremely useful in Data Science and Machine Learning when performing Exploratory Data Analysis (EDA).

---

# Why Subplots are Important?

Subplots help answer questions such as:

* How do multiple features behave?
* How can different visualizations be compared?
* How can several charts be displayed in one report?

Data Scientists often use subplots to create dashboards and EDA reports.

---

# Real-World Examples

### Employee Analysis

Display:

* Salary Distribution
* Experience Distribution
* Department Count
* Salary Trend

in a single figure.

---

### Machine Learning Analysis

Display:

* Histogram
* Box Plot
* Scatter Plot
* Correlation Plot

together.

---

# Subplot Syntax

```python
plt.subplot(rows, columns, position)
```

Example:

```python
plt.subplot(2, 2, 1)
```

Meaning:

```text
2 Rows
2 Columns
Position 1
```

---

# Understanding Positions

```text
2 x 2 Grid

┌─────┬─────┐
│  1  │  2  │
├─────┼─────┤
│  3  │  4  │
└─────┴─────┘
```

Each number represents a subplot position.

---

# Why Use Subplots?

Instead of:

```text
Histogram → Window 1

Scatter Plot → Window 2

Bar Plot → Window 3

Line Plot → Window 4
```

We can display all charts together.

---




# Practical Example 1: Two Plots


````python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)

plt.hist([10,20,30,40,50])

plt.title("Histogram")

plt.subplot(1,2,2)

plt.plot([1,2,3,4,5],
         [10,20,30,40,50])

plt.title("Line Plot")

plt.tight_layout()

plt.show()
````

# Practical Example 2: 2×2 Subplots

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

# Histogram
plt.subplot(2,2,1)
plt.hist([45,50,55,60,65,70])
plt.title("Histogram")

# Scatter Plot
plt.subplot(2,2,2)
plt.scatter(   [1,2,3,4,5],
    [10,20,30,40,50]
)
plt.title("Scatter Plot")

# Bar Plot
plt.subplot(2,2,3)
plt.bar(
  ["A","B","C"],
    [10,15,20]
)
plt.title("Bar Plot")

# Line Plot
plt.subplot(2,2,4)
plt.plot(
    [1,2,3,4,5],
    [5,10,15,20,25]
)
plt.title("Line Plot")

plt.tight_layout()

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

salary = np.random.randint(
    25000,
    100000,
    100
)

experience = np.random.randint(
    1,
    15,
    100
)

plt.figure(figsize=(10,8))

# Histogram
plt.subplot(2,2,1)
plt.hist(salary)
plt.title("Salary Distribution")

# Box Plot
plt.subplot(2,2,2)
plt.boxplot(salary)
plt.title("Salary Box Plot")

# Scatter Plot
plt.subplot(2,2,3)
plt.scatter(experience, salary)
plt.title("Experience vs Salary")

# Line Plot
plt.subplot(2,2,4)
plt.plot(sorted(salary))
plt.title("Salary Trend")

plt.tight_layout()

plt.show()
````
