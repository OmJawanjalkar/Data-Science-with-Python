# Heatmap

## Introduction

A Heatmap is a graphical representation of data where values are represented using colors.

Heatmaps help us quickly identify:

* Relationships between variables
* Correlations
* Patterns in data
* Strong and weak relationships

In Data Science and Machine Learning, Heatmaps are mainly used to visualize the **correlation matrix**.

---

# Why Heatmap is Important?

Before building a Machine Learning model, we need to understand:

* Which features are strongly related?
* Which features are weakly related?
* Are there highly correlated features?
* Which features may be important for prediction?

Heatmaps provide these insights visually.

# Understanding Correlation

Correlation measures the relationship between two variables.

Range:

```text
-1  → Perfect Negative Correlation

 0  → No Correlation

+1  → Perfect Positive Correlation
```

Examples:

```text
Experience ↑
Salary ↑

Positive Correlation
```

```text
Price ↑
Demand ↓

Negative Correlation
```

# Correlation Matrix Example

````python
import pandas as pd

data = {
    "Experience": [1,2,3,4,5],
    "Salary": [25000,35000,45000,55000,65000],
    "Performance": [60,70,80,85,95]
}

df = pd.DataFrame(data)

print(df.corr())
````

---

# What Does the Color Mean?

```text
Dark Color  → Strong Relationship

Light Color → Weak Relationship
```

Higher correlation values usually appear darker.

---

# Creating a Heatmap Using Matplotlib


````python
import pandas as pd 
import matplotlib.pyplot as plt 

data = { "Experience": [1,2,3,4,5], 
        "Salary": [25000,35000,45000,55000,65000],
        "Performance": [60,70,80,85,95] 
        }

df = pd.DataFrame(data)

coorelation = df.corr()
plt.imshow(coorelation)
plt.colorbar() 
plt.show()
````

````python
# Heatmap with Labels


import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Experience": [1,2,3,4,5],
    "Salary": [25000,35000,45000,55000,65000],
    "Performance": [60,70,80,85,95]
}

df = pd.DataFrame(data)

correlation = df.corr()

plt.imshow(correlation)

plt.colorbar()

plt.xticks(
    range(len(correlation.columns)),
    correlation.columns
)

plt.yticks(
    range(len(correlation.columns)),
    correlation.columns
)

plt.title("Correlation Heatmap")

plt.show()
````
