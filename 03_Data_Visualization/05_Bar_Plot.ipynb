# Bar Plot

## Introduction

A Bar Plot is a chart used to compare different categories using rectangular bars.

The height (or length) of each bar represents the value of a category.

Bar Plots are one of the most commonly used visualizations in Data Science for comparing categorical data.

---

# Why Bar Plot is Important?

Bar Plots help answer questions such as:

* Which department has the highest salary?
* Which product has the highest sales?
* Which category has the most records?
* Which city has the highest population?

Bar Plots make comparisons easy and visually appealing.

---

# Real-World Examples

### Product Sales

Compare sales of different products.

### Student Performance

Compare marks of students.

### Department Analysis

Compare number of employees in each department.

### Population Analysis

Compare population across cities.

---

# Bar Plot Syntax

```python
plt.bar(x, y)
```

Where:

```text
x → Categories

y → Values
```

---

# Basic Bar Plot Example

````python
import matplotlib.pyplot as plt

students = ["A", "B", "C", "D"]

marks = [85, 90, 75, 95]

plt.bar(students, marks)

plt.show()
````

# Bar Plot with Title and Labels

````python
import matplotlib.pyplot as plt

students = ["A", "B", "C", "D"]

marks = [85, 90, 75, 95]

plt.bar(students, marks)

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.show()
````

# Horizontal Bar Plot

````python
import matplotlib.pyplot as plt

students = ["A", "B", "C", "D"]

marks = [85, 90, 75, 95]

plt.barh(students, marks)

plt.show()
````

# Example: Department Employee Count


````python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Department": ["IT", "HR", "Finance", "Sales"],
    "Employees": [50, 30, 25, 40]
}

df = pd.DataFrame(data)

plt.bar(
    df["Department"],
    df["Employees"]
)

plt.title("Employees by Department")

plt.xlabel("Department")

plt.ylabel("Number of Employees")

plt.show()
````

# Interview Questions

### 1. What is a Bar Plot?

---

### 2. When should we use a Bar Plot?

When comparing values across categories.

---

### 3. What is the difference between Histogram and Bar Plot?

Histogram:

* Numerical Data
* Continuous Data

Bar Plot:

* Categorical Data
* Discrete Categories

---

### 4. What does the height of a bar represent?

The value of the category.

---

### 5. What is the difference between bar() and barh()?

```text
bar()  → Vertical Bars

barh() → Horizontal Bars
```
