# Pie Chart

## Introduction

A Pie Chart is a circular chart used to represent data as parts of a whole.

Each slice of the pie represents a category's contribution to the total.

Pie Charts are useful when showing percentages or proportions.

---

# Why Pie Chart is Important?

Pie Charts help answer questions such as:

* What percentage of employees belong to each department?
* What is the market share of each company?
* Which product contributes the most sales?
* How is the budget distributed?

---

# Real-World Examples

### Market Share

* Apple → 40%
* Samsung → 30%
* Xiaomi → 20%
* Others → 10%

### Department Employees

* IT → 40%
* HR → 20%
* Finance → 15%
* Sales → 25%

### Product Sales Distribution

Different products contributing to total sales.

---

# Understanding Pie Chart

Each slice represents:

```text
Category Contribution / Total Contribution
```

Larger slice:

```text
Higher Contribution
```

Smaller slice:

```text
Lower Contribution
```

---

# Pie Chart Syntax

```python
plt.pie(values)
```

---

# Important Parameters

| Parameter  | Purpose          |
| ---------- | ---------------- |
| values     | Numerical values |
| labels     | Category names   |
| autopct    | Show percentages |
| startangle | Rotate chart     |
| title()    | Add title        |

---

# Practical Example 1: Basic Pie Chart

````python
import matplotlib.pyplot as plt

employees = [40, 20, 15, 25]

plt.pie(employees)

plt.show()
````

# Practical Example 2: Pie Chart with Labels and Percentages

````python
import matplotlib.pyplot as plt

departments = ["IT", "HR", "Finance", "Sales"]

employees = [40, 20, 15, 25]

plt.pie(
  employees,
  labels=departments,
  autopct="%1.1f%%"
)

plt.title("Department Employee Distribution")

plt.show()
````

# Real Data Science Example

````python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Department": ["IT", "HR", "Finance", "Sales"],
    "Employees": [50, 25, 15, 10]
}

df = pd.DataFrame(data)

plt.pie(
    df["Employees"],
    labels=df["Department"],
    autopct="%1.1f%%"
)

plt.title("Company Employee Distribution")

plt.show()
````

# When to Use Pie Chart?

Use Pie Charts when:

* Showing percentages
* Showing category contribution
* Comparing parts of a whole

Examples:

```text
Market Share

Department Distribution

Budget Allocation

Product Sales Share
```

---

# Advantages

* Easy to understand
* Shows proportions clearly
* Useful for business reports
* Effective for small number of categories

---

# Limitations

* Not suitable for many categories
* Difficult to compare close values
* Bar Charts are often better for precise comparison

---

# Applications in Data Science

* Category Distribution
* Business Analytics
* Sales Analysis
* Market Research
* Reporting Dashboards
