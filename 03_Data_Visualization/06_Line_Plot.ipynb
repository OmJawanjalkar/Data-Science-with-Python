# Line Plot

## Introduction

A Line Plot is a chart used to display data points connected by straight lines.

It is mainly used to visualize trends and changes over time.

Line plots help us understand how values increase, decrease, or fluctuate over a period.

---

# Why Line Plot is Important?

Line plots help answer questions such as:

* How are sales changing over time?
* Is company revenue increasing?
* How does stock price change daily?
* How does model accuracy improve during training?

Line plots are commonly used in Data Science, Business Analytics, and Machine Learning.

---

# Real-World Applications

### Sales Analysis

Track monthly sales trends.

### Stock Market Analysis

Monitor stock prices over time.

### Weather Analysis

Track temperature changes.

### Machine Learning

Monitor:

* Training Loss
* Validation Loss
* Model Accuracy

---

# Line Plot Syntax

```python
plt.plot(x, y)
```

Where:

```text
x → Time or ordered values

y → Numerical values
```


---

# Important Parameters

| Parameter | Purpose       |
| --------- | ------------- |
| x         | X-axis values |
| y         | Y-axis values |
| title()   | Add title     |
| xlabel()  | X-axis label  |
| ylabel()  | Y-axis label  |
| show()    | Display chart |

---

# Practical : Implemention  

````python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

sales = [10000, 12000, 15000, 13000, 18000, 20000]

plt.plot(months, sales)
plt.title("Monthly Sales Trends")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
````

# Practical Example 2: Machine Learning Training Accuracy

````python
import matplotlib.pyplot as plt

epochs = [1, 2, 3, 4, 5, 6, 7]

accuracy = [60, 68, 74, 80, 85, 89, 92]

plt.plot(epochs, accuracy)
plt.title("Model Accuracy vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.show()
````
