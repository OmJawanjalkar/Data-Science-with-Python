# NumPy Reshaping Arrays

## Introduction

Reshaping is the process of changing the shape or dimensions of an array without changing its data.

In Data Science and Machine Learning, data often needs to be converted into different formats before analysis or model training. NumPy provides several functions to reshape arrays efficiently.

---

# Why Do We Need Reshaping?

Real-world datasets are not always available in the required format.

For example:

```text
Original Data
[1, 2, 3, 4, 5, 6]
```

may need to be converted into:

```text
[[1, 2, 3],
 [4, 5, 6]]
```

for machine learning or data analysis.

---

# What is Reshaping?

Reshaping changes the structure of an array while keeping the same data.

Example:

```text
1D Array → 2D Array
2D Array → 1D Array
2D Array → 3D Array
```

---


# Important Reshaping Functions

## 1. reshape()

The `reshape()` function is used to change the shape of an array.



````python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

new_arr = arr.reshape(2, 3)

print(new_arr)
````

# 2. reshape(-1)

      NumPy automatically calculates the missing dimension.

````python
import numpy as np 

arr = np.array([1, 2, 3, 4, 5, 6])

new_arr = arr.reshape(-1, 3)

print(new_arr)
````

````python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

new_arr = arr.reshape(3, -1)

print(new_arr)
````

# 3. flatten()

        Converts a multi-dimensional array into a 1D array.


````python
import numpy as np

arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

new_arr = arr.flatten()

print(new_arr)
````

# 4. ravel()

      Converts a multi-dimensional array into a 1D array.

````python
import numpy as np

arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

ravel_arr = arr.ravel()


print(ravel_arr)
````

### Difference Between flatten() and ravel()

````python
import numpy as np

arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

flat_arr = arr.flatten()
ravel_arr = arr.ravel()

flat_arr[0] = 100
ravel_arr[1] = 200

print("Original Array")
print(arr)

print("\nFlatten Array")
print(flat_arr)

print("\nRavel Array")
print(ravel_arr)
````


---

# Difference Between flatten() and ravel()

| flatten()         | ravel()           |
| ----------------- | ----------------- |
| Returns a copy    | Returns a view    |
| More memory usage | Less memory usage |
| Slower            | Faster            |

---

# 5. transpose()

      The transpose() function swaps rows and columns.

      Example:

      arr.T

      or

      arr.transpose()

````python
import numpy as np

arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("Method 1")
print(arr. T)


print("Method 2")
print(arr.transpose())
````
