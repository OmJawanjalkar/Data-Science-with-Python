# Top 20 NumPy Interview Questions and Answers

## 1. What is NumPy?

**Answer:**

NumPy (Numerical Python) is a Python library used for numerical computations, array operations, mathematical functions, and scientific computing.

---

## 2. Why is NumPy faster than Python Lists?

**Answer:**

NumPy arrays are stored in contiguous memory locations and are implemented in C, making them faster and more memory efficient than Python lists.

---

## 3. What is a NumPy Array?

**Answer:**

A NumPy array is a collection of elements of the same data type stored in a contiguous memory block.

Example:

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
```

---

## 4. What is the difference between a Python List and a NumPy Array?

**Answer:**

| Python List          | NumPy Array                    |
| -------------------- | ------------------------------ |
| Slower               | Faster                         |
| More Memory          | Less Memory                    |
| Different Data Types | Same Data Type                 |
| Limited Operations   | Supports Vectorized Operations |

---

## 5. What is Vectorization?

**Answer:**

Vectorization means performing operations on entire arrays without using loops.

Example:

```python
arr * 2
```

---

## 6. What is Broadcasting?

**Answer:**

Broadcasting allows NumPy to perform operations on arrays of different shapes without explicit looping.

Example:

```python
arr + 10
```

---

## 7. What is the difference between reshape() and resize()?

**Answer:**

* reshape() returns a new array.
* resize() modifies the original array.

---

## 8. What does shape attribute return?

**Answer:**

It returns the dimensions of an array.

Example:

```python
arr.shape
```

Output:

```text
(rows, columns)
```

---

## 9. What does ndim return?

**Answer:**

It returns the number of dimensions of an array.

Example:

```python
arr.ndim
```

---

## 10. What does dtype return?

**Answer:**

It returns the data type of array elements.

Example:

```python
arr.dtype
```

---

## 11. What is Boolean Masking?

**Answer:**

Boolean Masking is a technique used to filter data using conditions.

Example:

```python
arr[arr > 50]
```

---

## 12. What is the difference between flatten() and ravel()?

**Answer:**

| flatten()                           | ravel()                       |
| ----------------------------------- | ----------------------------- |
| Returns Copy                        | Returns View                  |
| More Memory                         | Less Memory                   |
| Changes don't affect original array | Changes affect original array |

---

## 13. What is the purpose of np.random.seed()?

**Answer:**

It makes random outputs reproducible.

Example:

```python
np.random.seed(42)
```

---

## 14. What is the difference between rand() and randn()?

**Answer:**

### rand()

Generates random values between 0 and 1.

### randn()

Generates random values from a normal distribution.

---

## 15. What is the difference between np.dot() and np.matmul()?

**Answer:**

### np.dot()

Computes dot product.

### np.matmul()

Performs matrix multiplication.

---

## 16. What is the transpose of a matrix?

**Answer:**

Transpose swaps rows and columns.

Example:

```python
arr.T
```

---

## 17. What is a determinant?

**Answer:**

A determinant is a scalar value used to determine whether a matrix is invertible.

Example:

```python
np.linalg.det(A)
```

---

## 18. What is the purpose of np.mean(), np.median(), and np.std()?

**Answer:**

* mean() → Average Value
* median() → Middle Value
* std() → Standard Deviation

---

## 19. What is the difference between axis=0 and axis=1?

**Answer:**

### axis=0

Column-wise operation.

### axis=1

Row-wise operation.

Example:

```python
np.sum(arr, axis=0)
np.sum(arr, axis=1)
```

---

## 20. Why is NumPy important in Data Science?

**Answer:**

NumPy provides efficient array operations, mathematical functions, broadcasting, and linear algebra tools that form the foundation of Pandas, Scikit-Learn, TensorFlow, and other Data Science libraries.
