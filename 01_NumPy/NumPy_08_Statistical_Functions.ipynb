# Introduction

### Statistical functions are used to summarize, analyze, and understand data.

In Data Science, statistical functions help us answer questions such as:

What is the average value?
What is the highest value?
What is the lowest value?
How spread out is the data?

NumPy provides built-in statistical functions that perform these calculations efficiently.

# Important Statistical Functions

### 1. sum()

    Returns the total sum of all elements.  eg. np.sum()

````python
import numpy as np

marks = np.array([80, 90, 75, 95, 88])

print(marks)
````

````python
print(np.sum(marks))
````

### 2. mean()

    Returns the average value.

````python
print(np.mean(marks))
````

### 3. median()

      Returns the middle value after sorting the data.

````python
print(np.median(marks))
````

### 4. min()

     Returns the smallest value.  eg.np.min(arr)


````python
print(np.min(marks))
````

### 5. max()

   Returns the largest value.  eg. np.max(arr)


````python
print(np.max(marks))
````

### 6. std()

    Returns the standard deviation.   eg. np.std(arr)

    Standard deviation measures how much data varies from the mean.


````python
print(np.std(marks))
````

### 7. var()

      Returns the variance. eg. np.var(arr)

      Variance measures the spread of data.


````python
print(np.var(marks))
````


---

# Axis Parameter

The axis parameter allows calculations row-wise or column-wise.

### axis = 0

Column-wise operation.

### axis = 1

Row-wise operation.

````python
import numpy as np

marks = np.array([
    [80, 90, 85],
    [70, 88, 92],
    [95, 91, 87]
])

print("Column-wise Sum")
print(np.sum(marks, axis=0))

print("\nRow-wise Sum")
print(np.sum(marks, axis=1))
````


---
### eg. Real Data Science Example 

````python
import numpy as np

sales = np.array([
    10000,
    12000,
    15000,
    18000,
    17000
])

print("Total Sales:", np.sum(sales))
print("Average Sales:", np.mean(sales))
print("Highest Sales:", np.max(sales))
print("Lowest Sales:", np.min(sales))
````
