# NumPy Array Attributes

## Introduction

After creating a NumPy array, it is important to understand its structure and properties.

NumPy provides several built-in attributes that help us understand the shape, dimensions, size, and data type of an array.

These attributes are widely used in Data Science, Machine Learning, and Deep Learning because they help us inspect and validate datasets before performing analysis.

---

### Practical Implementation 1: Student Marks Dataset

````python
import numpy as np

student_marks = np.array([ [80, 90, 85],
    [70, 88, 92],
    [95, 91, 87],
    [78, 82, 89]
])

print(student_marks)
````

### shape  → How many rows and columns?

````python
print("Shape :", student_marks.shape)
````

### ndim   → How many dimensions?

````python
print("Dimension :", student_marks.ndim)
````

### size   → How many total values?

````python
print("Size :", student_marks.size)
````

### dtype  → What type of data is stored?

````python
print("Data Type :", student_marks.dtype)
````



# Why Are Array Attributes Important?

Before working with any dataset, a Data Scientist needs answers to questions such as:

* How many rows and columns are present?
* How many dimensions does the data have?
* How many values are stored in the dataset?
* What is the data type of the values?

NumPy attributes provide this information quickly and efficiently.

---

# Important Array Attributes

NumPy provides four fundamental attributes:

```python
shape
ndim
size
dtype
```

These attributes are frequently used while performing data analysis and machine learning tasks.

---

# 1. shape

## Definition

The `shape` attribute returns the dimensions of an array.

For a two-dimensional array, it returns:

```text
(Number of Rows, Number of Columns)
```

### Example

```python
arr.shape
```

### Use Cases

* Understanding dataset structure
* Checking rows and columns
* Verifying data before model training

### Data Science Example

Suppose a dataset contains:

```text
1000 Students
5 Features
```

The shape will be:

```text
(1000, 5)
```

This means:

* 1000 rows (records)
* 5 columns (features)

---

# 2. ndim

## Definition

The `ndim` attribute returns the number of dimensions of an array.

### Types of Arrays

### One-Dimensional Array

```text
[10, 20, 30]
```

Dimension:

```text
1
```

---

### Two-Dimensional Array

```text
[
 [1, 2, 3],
 [4, 5, 6]
]
```

Dimension:

```text
2
```

---

### Three-Dimensional Array

Used in image processing and deep learning.

Dimension:

```text
3
```

---

## Use Cases

* Understanding data structure
* Working with images and tensors
* Preparing data for machine learning models

---

# 3. size

## Definition

The `size` attribute returns the total number of elements present in an array.

### Formula

```text
Rows × Columns
```

Example:

```text
Shape = (3, 4)

Size = 12
```

---

## Use Cases

* Counting observations
* Dataset validation
* Memory estimation

---

### Data Science Example

If a dataset contains:

```text
500 rows
8 columns
```

Total elements:

```text
500 × 8 = 4000
```

---

# 4. dtype

## Definition

The `dtype` attribute returns the data type of the elements stored in an array.

Examples:

```text
int64
float64
bool
str
```

---

## Why dtype is Important?

Machine Learning algorithms work primarily with numerical data.

Before training a model, it is important to ensure that the data types are correct.

---

### Examples

```text
Age      → Integer
Salary   → Float
Gender   → String
Purchased → Boolean
```

Checking `dtype` helps identify incorrect data types and data quality issues.
