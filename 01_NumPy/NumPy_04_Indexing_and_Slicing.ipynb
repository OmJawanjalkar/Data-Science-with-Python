# NumPy Indexing and Slicing

## Introduction

Indexing and slicing are used to access specific elements or a group of elements from a NumPy array.

These operations are important because in Data Science we rarely work with an entire dataset. Instead, we often select specific rows, columns, or values for analysis.

---

# What is Indexing?

Indexing is used to access a single element from an array.

Example:

```python
arr[0]
```

Indexing returns one value.

---

# What is Slicing?

Slicing is used to access multiple elements from an array.

Example:

```python
arr[1:5]
```

Slicing returns a subset of the array.

---

# Why is Indexing and Slicing Important?

Used for:

* Data Selection
* Data Cleaning
* Feature Selection
* Machine Learning
* Data Analysis

---


# Practical Implementation
### 1. Creating Sample Array

````python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print(arr)
````

# Types of Indexing and Slicing
### 2. 1D Array Indexing

      Access individual elements.

````python
print(arr[0])
print(arr[2])
print(arr[-1])
print(arr[-4])
````

### 3. 1D Array Slicing

````python
print(arr[1:3])
print(arr[0:5])
print(arr[3:])
print(arr[:3])
````

### 4. Creating a 2D Array

````python
data = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print(data)
````

### 5. 2D Array Indexing

````python
print(data[0, 0])
print(data[2, 2])
print(data[2, 2])
print(data[1, 2])
print(data[2, 1])
````

### 6. Selecting Rows

````python
print(data[0])
print(data[2])
print(data[1])
````

### 7. Selecting Columns

````python
print(data[:, 0])

print(data[:, 1])

print(data[:, 2])
````

### 8. 2D Array Slicing

````python
print(data[0:2, 0:2])
````

### 9. Fancy Indexing

````python
arr = np.array([10, 20, 30, 40, 50])

print(arr[[0, 3, 4]])
````

### 10. Boolean Indexing

````python
arr = np.array([10, 20, 30, 40, 50])

print(arr[arr > 25])
````

---

## 2. 2D Array Indexing

Access elements using:

```python
arr[row, column]
```

---

## 3. 1D Array Slicing

Access a range of elements.

```python
arr[start:end]
```

---

## 4. 2D Array Slicing

Access specific rows and columns.

```python
arr[row_start:row_end,
    col_start:col_end]
```

---

---

## 6. Boolean Indexing

Select elements based on conditions.

Most important in Data Science.
