# Introduction to NumPy

## What is NumPy?

NumPy stands for **Numerical Python**.

It is a powerful Python library used for numerical computing, scientific computing, and data analysis.

NumPy provides support for:

* Multi-dimensional arrays
* Mathematical operations
* Statistical functions
* Linear algebra
* Random number generation

---

# Why Do We Need NumPy?

Python lists are flexible but slow when working with large amounts of numerical data.

NumPy is designed to perform calculations much faster and more efficiently.

An array is a collection of similar data types stored in contiguous memory locations.

Example:

```python

```

# NumPy array

````python
import numpy as np
````

````python
numbers = np.array([1, 2, 3, 4, 5])

result = numbers*2

print(result)
````

# Why is NumPy Faster?

### 1. Contiguous Memory

NumPy stores data in continuous memory blocks.

```text
10 20 30 40 50
```

This improves performance.

---

### 2. Vectorization

Instead of using loops:

```python
for i in range(len(arr)):
    arr[i] *= 2
```

NumPy performs operations on the entire array at once.

```python
arr * 2
```

---

### 3. Optimized C Implementation

Most NumPy operations are implemented in C, making them much faster than pure Python loops.
