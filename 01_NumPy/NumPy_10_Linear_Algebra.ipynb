# NumPy Linear Algebra

## Introduction

Linear Algebra is one of the most important mathematical foundations of Data Science, Machine Learning, Deep Learning, and Artificial Intelligence.

Many machine learning algorithms use matrices and vectors internally to perform calculations and make predictions.

NumPy provides a powerful Linear Algebra module that allows us to perform matrix operations efficiently.

---

Linear Algebra is used in:

* Machine Learning
* Deep Learning
* Computer Vision
* Natural Language Processing
* Recommendation Systems

Understanding these concepts helps in understanding how machine learning models work internally.

---

### 1. Vectors and Matrices

````python
import numpy as np

# Vector
v = np.array([1, 2, 3])

# Matrix
A = np.array([[1, 2],
              [3, 4]])
````

### 2. Matrix Operations

Addition / Subtraction

````python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = A + B

print(result)
````

````python
result = B - A

print(result)
````

Matrix Multiplication

````python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = A @ B

print(result)
````

Element-wise multiplication

````python
A*B
````

3. Transpose

````python
A = np.array([[1, 2, 3],
              [4, 5, 6]])

print(A.T)
````

4. Determinant

````python
A = np.array([[1, 2],
              [3, 4]])

det = np.linalg.det(A)

print(det)
````

5. Inverse

````python
A = np.array([[1, 2],
              [3, 4]])

inverse = np.linalg.inv(A)

print(inverse)
````

````python
print(A @ inverse)
````

### 8. Solving Linear Equations

Suppose:

        2x+y=5
        x+3y=6

````python
A = np.array([[2, 1],
              [1, 3]])

B = np.array([5, 6])

solution = np.linalg.solve(A, B)

print(solution)
````

### 9. Eigenvalues and Eigenvectors

   Used in PCA, Machine Learning, Computer Vision, etc.

````python
A = np.array([[4, 2],
              [1, 3]])

eigenvalues, eigenvectors = np.linalg.eig(A)
````

````python
print(eigenvalues)
````

````python
print(eigenvectors)
````

### 10. Vector Norm

   Norm represents vector length.

````python
v = np.array([3, 4])

print(np.linalg.norm(v))
````

### 11. Rank of Matrix

Rank tells the number of independent rows or columns.

````python
A = np.array([[1, 2],
              [2, 4]])

print(np.linalg.matrix_rank(A))
````

| Function                  | Description                |
| ------------------------- | -------------------------- |
| `np.linalg.det()`         | Determinant                |
| `np.linalg.inv()`         | Matrix Inverse             |
| `np.linalg.solve()`       | Solve Linear Equations     |
| `np.linalg.eig()`         | Eigenvalues & Eigenvectors |
| `np.linalg.norm()`        | Vector/Matrix Norm         |
| `np.linalg.matrix_rank()` | Matrix Rank                |
| `A.T`                     | Transpose                  |
| `A @ B`                   | Matrix Multiplication      |

# Interview Questions

## 1. What is Linear Algebra in NumPy?

**Answer:**

Linear Algebra in NumPy is performed using the `numpy.linalg` module, which provides functions for matrix and vector operations such as matrix multiplication, determinant, inverse, eigenvalues, and solving linear equations.

---

## 2. What is the difference between `*` and `@`?

**Answer:**

### `*` Operator

Performs element-wise multiplication.

Example:

```python
A * B
```

### `@` Operator

Performs matrix multiplication.

Example:

```python
A @ B
```

---

## 3. What is a determinant?

**Answer:**

A determinant is a scalar value calculated from a square matrix that indicates whether the matrix is invertible.

Example:

```python
np.linalg.det(A)
```

---

## 4. When does a matrix have an inverse?

**Answer:**

A matrix has an inverse only when its determinant is not equal to zero.

```text
det(A) ≠ 0
```

If:

```text
det(A) = 0
```

the matrix is called a singular matrix and cannot be inverted.

---

## 5. Why is `np.linalg.solve()` preferred over `np.linalg.inv()`?

**Answer:**

`np.linalg.solve()` is preferred because:

* Faster
* More accurate
* Numerically stable
* Uses less computation

It directly solves:

```text
AX = B
```

without explicitly calculating the inverse matrix.

---

## 6. What are Eigenvalues and Eigenvectors?

**Answer:**

Eigenvalues and Eigenvectors describe important properties of matrix transformations.

They are widely used in:

* Principal Component Analysis (PCA)
* Machine Learning
* Computer Vision
* Dimensionality Reduction

Example:

```python
np.linalg.eig(A)
```

---

## 7. What is the norm of a vector?

**Answer:**

The norm represents the length or magnitude of a vector.

Example:

```python
np.linalg.norm(v)
```

Applications:

* Distance calculations
* Similarity measurement
* Machine Learning

---

## 8. What is matrix rank?

**Answer:**

Matrix rank is the number of linearly independent rows or columns in a matrix.

Example:

```python
np.linalg.matrix_rank(A)
```

Rank helps determine:

* Data redundancy
* Matrix invertibility
* Linear dependence

---

## 9. What is the transpose of a matrix?

**Answer:**

The transpose of a matrix is obtained by interchanging rows and columns.

Example:

```python
A.T
```

If:

```text
[[1 2 3]
 [4 5 6]]
```

Transpose:

```text
[[1 4]
 [2 5]
 [3 6]]
```
