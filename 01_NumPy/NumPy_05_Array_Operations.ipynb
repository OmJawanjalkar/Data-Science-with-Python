# NumPy Array Operations

## Introduction

NumPy allows mathematical operations to be performed directly on arrays.

Unlike Python lists, NumPy can perform operations on all elements of an array simultaneously without using loops.

This feature is known as Vectorization.

---

# Why Array Operations Are Important?

Array operations are widely used in:

* Data Analysis
* Machine Learning
* Deep Learning
* Scientific Computing
* Feature Engineering

---

# What is Vectorization?

Vectorization means applying an operation to an entire array at once.

Example:

```python
arr * 2
```

Instead of using a loop, NumPy performs the operation on all elements efficiently.

---


# Types of Array Operations

## Arithmetic Operations

* Addition (+)
* Subtraction (-)
* Multiplication (*)
* Division (/)
* Power (**)

---

````python
import numpy as np

arr1 = np.array([10, 20, 30, 40, 50])

arr2 = np.array([1, 2, 3, 4, 5])

print(arr1)
print(arr2)
````

````python
print("Addition of arr :",arr1 + arr2)
print("Substraction :", arr1 - arr2)
print("Multipliction:" ,arr1 * arr2)
print("Division:", arr1 / arr2)
````

###  Power Operation

````python
print(arr1 ** 2)
````

###  Scalar Operations

````python
# ADD 10 

print(arr1 + 10)
````

Multiply by 2

````python
print(arr1 * 2)
````

## Comparison Operations

* Greater Than (>)
* Less Than (<)
* Equal To (==)
* Not Equal To (!=)

---

````python
import numpy as np

marks = np.array([35, 45, 78, 25, 90])

# Greater Than
print(marks > 50)

# Less Than
print(marks < 50)

# Greater Than or Equal To
print(marks >= 40)

# Less Than or Equal To
print(marks <= 40)

# Equal To
print(marks == 78)

# Not Equal To
print(marks != 78)

# Values Greater Than 50
print(marks[marks > 50])

# Values Less Than 50
print(marks[marks < 50])

# Values Between 40 and 80
print(marks[(marks >= 40) & (marks <= 80)])

# Values Greater Than 80 or Less Than 30
print(marks[(marks > 80) | (marks < 30)])

# Passed Students (Marks >= 40)
print(marks[marks >= 40])

# Failed Students (Marks < 40)
print(marks[marks < 40])
````

## Aggregation Operations

* sum()
* mean()
* min()
* max()
* std()

These operations summarize data and are heavily used in Data Science.

---

````python
import numpy as np

marks = np.array([80, 90, 75, 95, 88])
````

````python
print("Total mark :", np.sum(marks))
print("Avrage marks : ", np.mean(marks))
print("Lowest mark mark :", np.max(marks))
print("Higest mark :", np.max(marks))
print("Standard Deviation :", np.std(marks))
````

# Real Business Example
      
      Monthly Sales

````python
sales = np.array([
    10000,
    12000,
    15000,
    18000,
    17000
])

print("Total Sales :", np.sum(sales))
print("Average Sales :", np.mean(sales))
print("Maximum Sales :", np.max(sales))
````
