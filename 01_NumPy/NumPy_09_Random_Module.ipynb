# NumPy Random Module

## Introduction

The NumPy Random Module is used to generate random numbers, random arrays, and random samples.

Random data is widely used in Data Science, Machine Learning, Deep Learning, simulations, and testing.

NumPy provides a collection of functions that make random number generation easy and efficient.

---

# Why Do We Need Random Numbers?

Random numbers are useful for:

* Creating sample datasets
* Data simulation
* Testing algorithms
* Machine Learning experiments
* Train-Test data splitting
* Initializing model parameters

---

# Important Random Functions

NumPy provides several functions for generating random values:

```python
np.random.rand()
np.random.randn()
np.random.randint()
np.random.choice()
np.random.seed()
```

---

# 1. np.random.rand()

Generates random floating-point numbers between:

```text
0 and 1
```

The values are uniformly distributed.

---

````python
import numpy as np
````

````python
# Example 

print(np.random.rand(5))
````


 ---
 
# 2. np.random.randn()

Generates random values from a normal distribution.

Characteristics:

* Mean ≈ 0
* Standard Deviation ≈ 1

This function is commonly used in Machine Learning and Deep Learning.

---

````python
print(np.random.randn())
````

# 3. np.random.randint()

Generates random integer values within a specified range.

Example:

````python
print(np.random.randint(1, 100, 5))
````

# 4. np.random.choice()

Selects random values from a given list or array.

Example:

````python
list = ["Apple", "Pinapple", "Mongo", "Banana", "kivi", "Orange"]

print(np.random.choice(list))
````

# 5. np.random.seed()

The seed function ensures reproducibility.

Example:

````python
import numpy as np

np.random.seed(56)

print(np.random.randint(1, 100, 5))
````

# Combined Program

````python
import numpy as np

np.random.seed(42)

print("Random Decimal Numbers")
print(np.random.rand(5))

print("\nRandom Numbers from Normal Distribution")
print(np.random.randn(5))

print("\nRandom Integers")
print(np.random.randint(1, 100, 5))

print("\nRandom Choice")
print(np.random.choice(["Python", "NumPy", "Pandas", "Machine Learning"]))
````


---

# Real Data Science Example

Generate Student Marks Dataset

````python
import numpy as np

np.random.seed(42)

marks = np.random.randint(35, 100, 5)

print(marks)
````

````python
print("Average Marks :", np.mean(marks))
print("Highest Marks :", np.max(marks))
print("Lowest Marks :", np.min(marks))
````

### rand()     → Random Decimal Numbers (0 to 1)

### randn()    → Random Numbers from Normal Distribution

### randint()  → Random Integers

### choice()   → Random Selection

### seed()     → Same Output Every Run
