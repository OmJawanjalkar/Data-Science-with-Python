# NumPy Array Creation
### Introduction

Before performing any operation in NumPy, we first need to create arrays.

NumPy provides multiple functions to create arrays efficiently.

### import numpy as np

````python
import numpy as np
````

### 1. np.array()

      Used to create a NumPy array from a Python list.

````python
import numpy as np 

arr = np.array([10, 20, 30, 40, 50])

print(arr)
````

### Creating a 2D Array

````python
import numpy as np

arr = np.array([[1, 2, 3],[4, 5, 6]])

print(arr)
````

### 2. np.zeros()

      Creates an array filled with zeros.

````python
arr = np.zeros(5)
print(arr)
````

### 2D Ones Array

````python
arr = np.ones((2, 3))
print(arr)
````

### 4. np.arange()

     Creates an array within a range.

Syntax
          np.arange(start, stop, step)

````python
# Example 1

arr = np.arange(5)
print(arr)
````

````python
# Example 2

arr = np.arange(1, 10)
print(arr)
````

````python
# Example 3

arr = np.arange(1, 22, 2)
print(arr)
````

### 5. np.linspace()

      Creates evenly spaced values between two numbers.

### Syntax
      np.linspace(start, stop, number_of_values)

````python
# np.linspace() creates equally spaced numbers between a start value and an end value.

import numpy as np 

arr = np.linspace(1, 10, 5)
print(arr)
````

### 6. np.eye()

      Creates an Identity Matrix.

````python
import numpy as np

arr = np.eye(4)
print(arr)
````

# Difference Between arange() and linspace()

| arange()                   | linspace()                  |
| -------------------------- | --------------------------- |
| Uses step size             | Uses number of values       |
| Stop value may be excluded | Stop value included         |
| Integer based              | Often used for float values |

---

## Example

```python
print(np.arange(1, 10, 2))

print(np.linspace(1, 10, 5))
```
