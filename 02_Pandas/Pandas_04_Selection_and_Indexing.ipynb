# Pandas Selection and Indexing

## Introduction

Selection and Indexing are used to access specific rows, columns, and values from a DataFrame.

In real-world Data Science projects, datasets may contain thousands or millions of records. Instead of working with the entire dataset, we often need to select only the required rows and columns.

Pandas provides several methods for selecting data efficiently.

---

# Sample Dataset

````python
import pandas as pd

employee_data = {
    "Employee_ID": [101,102,103,104,105,106,107,108,109,110],
    "Name": ["John","Alice","Bob","David","Emma",
             "Chris","Sophia","Michael","Olivia","Daniel"],
    "Department": ["IT","HR","Finance","IT","HR",
                   "Finance","IT","HR","Finance","IT"],
    "Age": [25,30,28,35,27,32,29,31,26,34],
    "Salary": [50000,45000,60000,70000,48000,
               65000,75000,52000,58000,72000],
    "Experience": [2,5,3,10,4,7,6,5,3,8]
}

df = pd.DataFrame(employee_data)

print(df)
````


---

# Why Selection and Indexing?

Used for:

* Selecting columns
* Selecting rows
* Filtering records
* Feature selection
* Data analysis
* Machine Learning preprocessing

# 1. Selecting a Single Column

````python
df["Name"]
````

````python
print(type(df["Name"]))
````


---

# 2. Selecting Multiple Columns


````python
print(df[["Name", "Salary"]])
````


---

# 3. Selecting Rows Using Slicing

````python
print(df[ : 5])
````

````python
print(df[2:7])
````


---

# 4. loc[] (Label-Based Indexing)

loc is used when selecting data using row labels and column names.

Syntax:

```python
df.loc[row, column]
```

````python
print(df.loc[3])
````


---

## Select Single Row

````python
print(df.loc[0])
````


---

## Select Multiple Rows

````python
print(df.loc[[2, 4, 5]])
````


---

## Select Specific Rows and Columns

````python
print(df.loc[[1,3,5], ["Name", "Salary"]])
````

 
---

## Select Range of Rows

````python
print(df.loc[3:6])
````


---

# 5. iloc[] (Position-Based Indexing) 

iloc selects data using numerical positions.

Syntax:

```python
df.iloc[row_position, column_position]
```

---


---

## Select Single Row

````python
print(df.iloc[3])
````


---

## Select Multiple Rows

````python
print(df.iloc[[1, 3, 4]])
````


---

## Select Specific Rows and Columns


````python
print(df.iloc[[1,3,5], [1,4]])
````


---

## Select Range

````python
print(df.iloc[2:5])
````

Important:

iloc excludes the ending index.

---

# Difference Between loc and iloc

| loc                | iloc                |
| ------------------ | ------------------- |
| Label Based        | Position Based      |
| Uses column names  | Uses column numbers |
| End index included | End index excluded  |

Example:

```python
df.loc[0:3]
```

Returns:

```text
0,1,2,3
```

---

```python
df.iloc[0:3]
```

Returns:

```text
0,1,2
```

# Summary Table

| Method              | Purpose                    |
| ------------------- | -------------------------- |
| df["col"]           | Single Column              |
| df[["col1","col2"]] | Multiple Columns           |
| df[start:end]       | Row Slicing                |
| loc[]               | Label-Based Selection      |
| iloc[]              | Position-Based Selection   |
| set_index()         | Change Index               |
| loc[row,col]        | Specific Value             |
| iloc[row,col]       | Specific Value by Position |
