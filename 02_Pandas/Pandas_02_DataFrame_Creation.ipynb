# Pandas DataFrame Creation

## Introduction

A DataFrame is the most important data structure in Pandas.

It is a two-dimensional tabular data structure consisting of rows and columns, similar to an Excel spreadsheet or SQL table.

Almost every Data Science, Data Analytics, and Machine Learning project uses DataFrames to store and manipulate data.

---

# What is a DataFrame?

A DataFrame is a collection of data organized into rows and columns.

Example:

| Student_ID | Name  | Marks |
| ---------- | ----- | ----- |
| 101        | John  | 85    |
| 102        | Alice | 90    |
| 103        | Bob   | 78    |

In Pandas, this table is represented as a DataFrame.

---

# Why Use DataFrames?

DataFrames provide:

* Easy data storage
* Data manipulation
* Data filtering
* Statistical analysis
* Data cleaning
* Data visualization support

They are the foundation of most Data Science workflows.

---

# Characteristics of DataFrames

* Two-dimensional structure
* Rows and columns
* Supports multiple data types
* Labeled axes
* Easy indexing and filtering
* Handles missing values

---


# Ways to Create a DataFrame

Pandas allows DataFrames to be created from multiple sources.

## 1. Dictionary

```python
pd.DataFrame(dictionary)
```

Most commonly used method.

---

````python
import pandas as pd

student_data = {
   "Student_ID": [101, 102, 103],
    "Name": ["John", "Alice", "Bob"],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(student_data)
print(df)
````

## 2. List of Lists

```python
pd.DataFrame(data)
```

Useful for structured tabular data.

---

````python
import pandas as pd 

data = [
    [101, "John", 85],
    [102, "Alice", 90],
    [103, "Bob", 78]
]

df = pd.DataFrame(data, columns=["Student_Id", "Name", "Marks" ])

print(df)
````

## 3. List of Dictionaries

```python
pd.DataFrame(data)
```

Widely used when working with APIs and JSON data.

---

````python
data = [
    {"Name": "John", "Marks": 85},
    {"Name": "Alice", "Marks": 90},
    {"Name": "Bob", "Marks": 78}
]

df = pd.DataFrame(data)

print(df)
````

## 4. NumPy Arrays

```python
pd.DataFrame(array)
```

Useful when working with numerical datasets.

---

````python
import numpy as  np 
import pandas as pd 

data = np.array([
    [101, 85],
    [102, 90],
    [103, 78]
])

df = pd.DataFrame(data, columns=["Student", "Mark"])

print(df)
````
