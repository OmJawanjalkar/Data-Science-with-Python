# Pandas Filtering and Boolean Masking

Filtering is the process of selecting specific rows from a DataFrame based on one or more conditions.

Boolean Masking creates a series of `True` and `False` values. Rows with `True` are selected, while rows with `False` are ignored.

Filtering is commonly used in:

* Data Cleaning
* Data Analysis
* Exploratory Data Analysis (EDA)
* Machine Learning
* Feature Engineering

### Common Operators

| Operator | Meaning               |
| -------- | --------------------- |
| >        | Greater Than          |
| <        | Less Than             |
| >=       | Greater Than Equal To |
| <=       | Less Than Equal To    |
| ==       | Equal To              |
| !=       | Not Equal To          |
| &        | AND                   |
| |        | OR                    |

---



```python

import pandas as pd

employee_data = {
    "Employee_ID": [101,102,103,104,105,106,107,108],
    "Name": ["John","Alice","Bob","David",
             "Emma","Chris","Sophia","Michael"],
    "Department": ["IT","HR","Finance","IT",
                   "HR","Finance","IT","HR"],
    "Age": [25,30,28,35,27,32,29,31],
    "Salary": [50000,45000,60000,70000,
               48000,65000,75000,52000]
}

df = pd.DataFrame(employee_data)

print("Original Dataset")
print(df)

# Salary Greater Than 60000
print("\nSalary > 60000")
print(df[df["Salary"] > 60000])

# Department = IT
print("\nIT Employees")
print(df[df["Department"] == "IT"])

# Age Less Than 30
print("\nAge < 30")
print(df[df["Age"] < 30])

```

### Multiple Conditions (AND)

```python
# Multiple Conditions (AND)
print("IT Employees with Salary > 60000")
print(
    df[
        (df["Department"] == "IT")
        &
        (df["Salary"] > 60000)
    ]
)
```

### Multiple Conditions (OR)

```python

print("IT OR HR Employees")
print(
    df[
        (df["Department"] == "IT")
        |
        (df["Department"] == "HR")
    ]
)
```

### isin()

```python

print("IT and Finance Employees")
print(
    df[
        df["Department"].isin(
            ["IT", "Finance"]
        )
    ]
)

```

### between()

```python

print("Salary Between 50000 and 70000")
print(
    df[
        df["Salary"].between(50000, 70000)
    ]
)
```

### String Filtering

```python

print("Names Containing 'a'")
print(
    df[
        df["Name"].str.contains("a", case=False)
    ]
)
```
