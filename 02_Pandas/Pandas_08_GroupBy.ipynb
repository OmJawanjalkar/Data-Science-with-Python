# Pandas GroupBy

## Introduction

The GroupBy operation is used to split data into groups, perform calculations on each group, and combine the results.

It follows the concept:

```text
Split
  ↓
Apply
  ↓
Combine
```

GroupBy is one of the most powerful features in Pandas and is heavily used in:

* Data Analysis
* Business Analytics
* Data Science
* Machine Learning
* Reporting Dashboards

---

# Why Use GroupBy?

Suppose a company wants to know:

* Average salary by department
* Maximum salary by department
* Total salary by department
* Number of employees in each department

GroupBy makes these tasks easy.

---


### Sample Dataset

````python
import numpy as np 
import pandas as pd 

data = {
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR', 'Finance', 'IT'],
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F', 'G'],
    'Salary': [50000, 40000, 60000, 70000, 45000, 80000, 55000]
}

df = pd.DataFrame(data)

df
````

### 1. Group By Single Column

````python
print(df.groupby('Department'))
````

### 2. Sum of Salary by Department

````python
df.groupby('Department')['Salary'].sum()
````

### 3. Average Salary by Department

````python
df.groupby('Department')['Salary'].mean()
````

### 4. Maximum Salary by Department

````python
df.groupby('Department')['Salary'].max()
````

### 5. Minimum Salary by Department

````python
df.groupby('Department')['Salary'].min()
````

### 6. Count Employees in Each Department

````python
df.groupby('Department')['Employee'].count()
````

### 7. Multiple Aggregations

````python
df.groupby('Department')['Salary'].agg(
  ['sum', 'mean', 'min', 'max', 'count']
)
````

### 8. Group By Multiple Columns

````python
data = {
    'Department': ['IT','IT','HR','HR','IT'],
    'Gender': ['M','F','M','F','M'],
    'Salary': [50000,60000,40000,45000,55000]
}

df = pd.DataFrame(data)

df.groupby(['Department','Gender'])['Salary'].mean()
````

### 9. Using as_index=False

By default, grouped columns become the index.

````python
df.groupby('Department', as_index=False)['Salary'].sum()
````

#### 10. Using get_group()

````python
grouped  = df.groupby('Department')

grouped.get_group('IT')
````

### Real Data Science Example

     Suppose you have sales data:

````python
sales = {
    'Region': ['North','South','North','East','South'],
    'Sales': [1000,2000,1500,3000,2500]
}

df = pd.DataFrame(sales)

df.groupby('Region')['Sales'].sum()
````

## 4. What is agg() used for?

