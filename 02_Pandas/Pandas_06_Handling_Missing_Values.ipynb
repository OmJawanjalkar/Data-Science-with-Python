# Pandas Handling Missing Values

## Introduction

In real-world datasets, some values are often missing due to data entry errors, system failures, incomplete surveys, or data collection issues.

These missing values are called **Missing Values** and are represented as:

```python
NaN
```

NaN stands for:

```text
Not a Number
```

Handling missing values is an important step in Data Cleaning and Data Preprocessing because Machine Learning models cannot work properly with missing data.

---

# Why Handle Missing Values?

Missing values can:

* Reduce model accuracy
* Cause errors during analysis
* Produce incorrect results
* Affect machine learning performance

Therefore, missing values must be identified and handled before analysis.

---

### Sample Dataset

````python
import pandas as pd 
import numpy as np 

data = { "Name": ["John", "Alice", "Bob", "David", "Emma"],
        "Age": [25, np.nan, 28, 35, np.nan],
        "Salary": [50000, 60000, np.nan, 70000, 55000], 
        "Department": ["IT", "HR", "Finance", np.nan, "IT"] 
        
      }

df = pd.DataFrame(data)

df
````

### Checking Missing Values
isnull()

Used to identify missing values.

````python
df.isnull()
````

````python
df.notnull()
````

### Count Missing Values

````python
df.isnull().sum()
````

### Removing Missing Values
   dropna()

   Removes rows containing missing values.

````python
df.dropna()
````

### Remove Columns And Row with Missing Values


Axis Parameter

axis=0 → Rows

axis=1 → Columns

````python
df.dropna(axis = 0)
````

````python
df.dropna(axis = 1)
````

### fill with Constant Value

````python
df["Age"] = df["Age"].fillna(0)
df
````

### Fill with Mean

````python
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())

df
````

### Fill with Median

````python
df["Age"] = df["Age"].fillna(df["Age"].median)
df
````

### Fill with Mode


````python
df["Department"] = df["Department"].fillna(
    df["Department"].mode()[0]
)

df
````


---

# Mean vs Median vs Mode

| Method | Best Used For                |
| ------ | ---------------------------- |
| Mean   | Numerical Data               |
| Median | Numerical Data with Outliers |
| Mode   | Categorical Data             |

Example:

```text
Age         → Median

Salary      → Mean

Department  → Mode
```


---

# Real Data Science Example

```python
import pandas as pd

df = pd.read_csv("employees.csv")

print(df.isnull().sum())

df["Age"] = df["Age"].fillna(
    df["Age"].median()
)

df["Salary"] = df["Salary"].fillna(
    df["Salary"].mean()
)

df["Department"] = df["Department"].fillna(
    df["Department"].mode()[0]
)
```

This is a common workflow before:

* Data Analysis
* EDA
* Feature Engineering
* Machine Learning

# Interview Questions

## 1. What are Missing Values?

Missing values are unavailable data points represented as NaN in Pandas.

---

## 2. What does isnull() do?

It identifies missing values in a DataFrame.

---

## 3. What does dropna() do?

It removes rows or columns containing missing values.

---

## 4. What does fillna() do?

It replaces missing values with a specified value.

---

## 5. When should we use Mean, Median, and Mode?

### Mean

Used for numerical data.

### Median

Used for numerical data containing outliers.

### Mode

Used for categorical data.

---

## 6. What is the difference between isnull() and notnull()?

```text
isnull()  → Missing Values

notnull() → Non-Missing Values
```
