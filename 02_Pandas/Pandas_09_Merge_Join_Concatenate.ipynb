# Pandas Merge, Join and Concatenate

## Introduction

In real-world Data Science projects, data is often stored in multiple tables, CSV files, Excel sheets, databases, or APIs.

To perform analysis, we need to combine these datasets into a single DataFrame.

Pandas provides three important functions for combining data:

```python
merge()
join()
concat()
```

These functions help us combine, append, and integrate data from different sources.

---

# Why Do We Need Merge, Join, and Concatenate?

Suppose we have:

### Employee Information

| Employee_ID | Name  |
| ----------- | ----- |
| 101         | John  |
| 102         | Alice |
| 103         | Bob   |

### Salary Information

| Employee_ID | Salary |
| ----------- | ------ |
| 101         | 50000  |
| 102         | 60000  |
| 103         | 55000  |

To analyze employee salaries, we must combine both tables.

This is where Merge, Join, and Concatenate are used.

---

# Difference Between Merge, Join, and Concat

| Function | Purpose                                 |
| -------- | --------------------------------------- |
| merge()  | Combine DataFrames using common columns |
| join()   | Combine DataFrames using index          |
| concat() | Stack DataFrames together               |

---

# 1. merge()

The merge() function combines DataFrames using one or more common columns.

It is similar to SQL JOIN operations.

Syntax:

```python
pd.merge(df1, df2, on="column_name")
```

---

## Types of Merge

### Inner Join

Returns only matching records from both DataFrames.

```python
pd.merge(df1, df2, on="ID", how="inner")
```

---

### Left Join

Returns all rows from the left DataFrame and matching rows from the right DataFrame.

```python
pd.merge(df1, df2, on="ID", how="left")
```

---

### Right Join

Returns all rows from the right DataFrame and matching rows from the left DataFrame.

```python
pd.merge(df1, df2, on="ID", how="right")
```

---

### Outer Join

Returns all rows from both DataFrames.

```python
pd.merge(df1, df2, on="ID", how="outer")
```

---

# 2. join()

The join() function combines DataFrames using indexes.

Syntax:

```python
df1.join(df2)
```

Join is useful when both DataFrames have meaningful indexes.

---

# 3. concat()

The concat() function combines DataFrames by stacking them.

Syntax:

```python
pd.concat([df1, df2])
```

Concatenation can be performed:

* Row-wise
* Column-wise

---

## Row-wise Concatenation

Adds rows from one DataFrame to another.

```python
pd.concat([df1, df2])
```

---

## Column-wise Concatenation

Adds columns side by side.

```python
pd.concat([df1, df2], axis=1)
```

---

# Practical 1: Inner Join

````python
import pandas as pd

employees = pd.DataFrame({
    "Employee_ID": [101, 102, 103],
    "Name": ["John", "Alice", "Bob"]
})

salary = pd.DataFrame({
    "Employee_ID": [101, 102, 103],
    "Salary": [50000, 60000, 55000]
})

result = pd.merge(
    employees,
    salary,
    on="Employee_ID",
    how="inner"
)


print(result )
````

# Practical 2: Left Merge

````python
import pandas as pd

employees = pd.DataFrame({
    "Employee_ID": [101, 102, 103, 104],
    "Name": ["John", "Alice", "Bob", "David"]
})

salary = pd.DataFrame({
    "Employee_ID": [101, 102, 103],
    "Salary": [50000, 60000, 55000]
})

result = pd.merge(
    employees,
    salary,
    on="Employee_ID",
    how="left"
)

print(result)
````

# Practical 3: Right Merge

````python
import pandas as pd

employees = pd.DataFrame({
    "Employee_ID": [101, 102, 103],
    "Name": ["John", "Alice", "Bob"]
})

salary = pd.DataFrame({
    "Employee_ID": [101, 102, 103, 104],
    "Salary": [50000, 60000, 55000, 70000]
})

result = pd.merge(
  employees,
  salary,
  on = "Employee_ID",
  how = "right"
  
)

print(result)
````

# Practical 4: Outer Merge

````python
import pandas as pd

employees = pd.DataFrame({
    "Employee_ID": [101, 102, 103],
    "Name": ["John", "Alice", "Bob"]
})

salary = pd.DataFrame({
    "Employee_ID": [101, 102, 104],
    "Salary": [50000, 60000, 70000]
})

result = pd.merge(
  employees,
  salary,
  on = "Employee_ID",
  how = "outer"
)

print(result)
````

## Practical 5: Join Example

````python
import pandas as pd

df1 = pd.DataFrame({
    "Name": ["John", "Alice", "Bob"]
})

df2 = pd.DataFrame({
    "Salary": [50000, 60000, 55000]
})

result = df1.join(df2)

print(result)
````

# Practical 6: Row-wise Concatenation

````python
import pandas as pd

df1 = pd.DataFrame({
    "Name": ["John", "Alice"]
})

df2 = pd.DataFrame({
    "Name": ["Bob", "David"]
})

result = pd.concat(
    [df1, df2],
    ignore_index=True
)

print(result)
````

# Practical 7: Column-wise Concatenation

````python
import pandas as pd

df1 = pd.DataFrame({
    "Name": ["John", "Alice", "Bob"]
})

df2 = pd.DataFrame({
    "Salary": [50000, 60000, 55000]
})

result = pd.concat(
    [df1, df2],
    axis=1
)

print(result)
````

# Practical 8: Real Data Science Example
    
    Customer and Orders Analysis

````python
import pandas as pd

customers = pd.DataFrame({
    "Customer_ID": [1, 2, 3, 4],
    "Customer_Name": [
        "John",
        "Alice",
        "Bob",
        "David"
    ]
})

orders = pd.DataFrame({
    "Customer_ID": [1, 2, 3, 4],
    "Order_Value": [
        5000,
        7000,
        3000,
        9000
    ]
})

result = pd.merge(
    customers,
    orders,
    on="Customer_ID"
)

print(result)
````
