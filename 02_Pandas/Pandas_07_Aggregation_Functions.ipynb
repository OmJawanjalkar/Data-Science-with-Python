# Pandas Aggregation Functions

## Introduction

Aggregation Functions are used to perform calculations on a column or an entire dataset and return a summarized result.

These functions help Data Scientists understand data quickly by calculating statistics such as average, total, minimum, maximum, and count.

Aggregation functions are widely used in:

* Data Analysis
* Exploratory Data Analysis (EDA)
* Business Analytics
* Machine Learning

---

# Why Use Aggregation Functions?

Aggregation functions help answer questions such as:

* What is the average salary?
* What is the highest salary?
* What is the lowest salary?
* How many employees are there?
* How many unique departments exist?

These insights help in understanding datasets before building Machine Learning models.

---



# Practical Implementation

````python
import pandas as pd

employee_data = {
    "Employee_ID": [101,102,103,104,105],
    "Name": ["John","Alice","Bob","David","Emma"],
    "Age": [25,30,28,35,27],
    "Salary": [50000,60000,45000,70000,55000]
}

df = pd.DataFrame(employee_data)

print("Sum:", df["Salary"].sum())

print("Mean:", df["Salary"].mean())

print("Median:", df["Salary"].median())

print("Minimum:", df["Salary"].min())

print("Maximum:", df["Salary"].max())

print("Count:", df["Salary"].count())

print("Standard Deviation:", df["Salary"].std())

print("Variance:", df["Salary"].var())

print("Unique Ages:", df["Age"].unique())

print("Number of Unique Ages:", df["Age"].nunique())
````

# Real Data Science Example

```python
import pandas as pd

sales_data = {
    "Product": ["Laptop","Mobile","Laptop","Tablet","Mobile"],
    "Sales": [50000,30000,45000,25000,35000]
}

df = pd.DataFrame(sales_data)

print("Total Sales:", df["Sales"].sum())

print("Average Sales:", df["Sales"].mean())

print("Highest Sales:", df["Sales"].max())

print("Lowest Sales:", df["Sales"].min())
```

# Interview Questions

## 1. What is an Aggregation Function?

An aggregation function performs calculations on multiple values and returns a single summarized result.

---

## 2. What is the difference between mean() and median()?

### mean()

Returns the average value.

### median()

Returns the middle value after sorting.

---

## 3. What does count() return?

It returns the number of non-null values.

---

## 4. What is the difference between unique() and nunique()?

### unique()

Returns unique values.

### nunique()

Returns the count of unique values.

---

## 5. What does std() calculate?

It calculates the standard deviation, which measures data spread.

---

## 6. What does var() calculate?

It calculates the variance of the data.
