# What are DataFrame Attributes?

  DataFrame attributes and methods help us understand the structure, size, columns, data types, and statistical information of a dataset.

In every Data Science project, after loading data using:

   df = pd.read_csv("data.csv")

the first thing we do is inspect the dataset using DataFrame attributes.

# Sample Dataset


````python
import pandas as pd 

data = {
    "Name": ["John", "Alice", "Bob", "David", "Emma"],
    "Age": [25, 30, 22, 35, 28],
    "Salary": [50000, 60000, 45000, 70000, 55000]
}

df = pd.DataFrame(data)

print(df)
````

### 1. shape 

     Returns the number of rows and columns.

````python
print(df.shape)
````

### 2. columns

     Returns all column names.

````python
print(df.columns)
````

    Convert to list:

````python
print(df.columns.tolist())
````

### 3. index

    Returns row labels.

````python
print(df.index)
````

### 4. dtypes 

    Returns data type of each column.

````python
print(df.dtypes)
````

### 5. info() 

     Provides complete dataset information.

````python
print(df.info())
````

### 6. describe() 

      Generates statistical summary.

````python
print(df.describe())
````

| Metric | Meaning            |
| ------ | ------------------ |
| count  | Number of values   |
| mean   | Average            |
| std    | Standard deviation |
| min    | Minimum            |
| max    | Maximum            |
| 50%    | Median             |


### 7. head() 

Displays first 5 rows.

````python
print(df.head)
````

### 8. tail()

    Displays last 5 rows.

````python
print(df.tail())
````

### 9. ndim

    Returns number of dimensions.

````python
print(df.ndim)
````

### 10. size

     Returns total number of elements.

Formula:

     Rows × Columns

````python
print(df.size)
````

# Most Important Interview Questions

## 1. What does `shape` return?

**Answer:** Returns the number of rows and columns.

```python
df.shape
```

---

## 2. What is the difference between `size` and `shape`?

**Answer:**

```python
shape → (rows, columns)

size → rows × columns
```

---

## 3. Why is `info()` important?

**Answer:** It provides dataset structure, missing values, and data types.

---

## 4. What does `describe()` return?

**Answer:** Statistical summary of numerical columns.

---

## 5. What is the difference between `head()` and `tail()`?

**Answer:**

```python
head() → First Rows

tail() → Last Rows
```
