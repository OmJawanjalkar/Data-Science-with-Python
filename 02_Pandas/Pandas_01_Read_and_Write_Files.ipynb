# Pandas Read and Write Files

## Introduction

In Data Science and Machine Learning, data is usually stored in files such as CSV, Excel, JSON, and SQL databases.

Before analyzing data, we must first load it into Pandas. After processing and analysis, we often save the results back to a file.

Pandas provides powerful functions to read and write data efficiently.

# Why Read and Write Files?

Reading and writing files is one of the most common tasks in Data Science.

Applications:

* Loading datasets
* Saving cleaned data
* Sharing analysis results
* Exporting reports
* Data preprocessing pipelines

# Why CSV is Most Important?

CSV files are the most commonly used format in Data Science because:

* Easy to create
* Easy to share
* Supported by almost every tool
* Lightweight
* Fast to read

Most datasets from Kaggle and real-world projects are available in CSV format.
