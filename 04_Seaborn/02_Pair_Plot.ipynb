# Pair Plot

## Introduction

A Pair Plot is a visualization used to display relationships between multiple numerical variables in a dataset.

It automatically creates a grid of plots where:

* Scatter Plots show relationships between variables.
* Histograms show the distribution of individual variables.

Pair Plots are one of the most useful visualizations during Exploratory Data Analysis (EDA).

---

# Why Pair Plot is Important?

When working with datasets containing multiple features, it becomes difficult to create separate scatter plots for every combination.

Pair Plot automatically generates all possible feature combinations, helping us:

* Understand relationships between variables
* Identify correlations
* Detect outliers
* Analyze distributions
* Explore datasets quickly

---

# Understanding Pair Plot

Suppose a dataset contains:

```text
Age
Experience
Salary
Performance
```

Pair Plot will create:

```text
Age vs Experience

Age vs Salary

Age vs Performance

Experience vs Salary

Experience vs Performance

Salary vs Performance
```

along with individual distributions.

---

# Components of Pair Plot

## Diagonal Plots

The diagonal shows the distribution of each feature.

Example:

```text
Age Distribution

Salary Distribution

Experience Distribution
```

---

## Off-Diagonal Plots

The remaining plots show relationships between two variables.

Example:

```text
Experience vs Salary

Age vs Salary

Salary vs Performance
```

---

# Why Pair Plot is Useful?

Instead of creating many scatter plots manually:

```text
Scatter Plot 1

Scatter Plot 2

Scatter Plot 3

Scatter Plot 4
```

Pair Plot generates everything automatically with a single function.

---

# Pair Plot Syntax

```python
sns.pairplot(df)
```

---



## Practical 1: Basic Pair Plot

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Age": [22, 25, 28, 30, 35],
    "Experience": [1, 2, 4, 6, 8],
    "Salary": [25000, 35000, 50000, 70000, 90000]
}

df = pd.DataFrame(data)

sns.pairplot(df)

plt.show()
````

## Practical 2: Pair Plot with hue

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Age": [22,25,28,30,35,40],
    "Salary": [25000,35000,50000,70000,90000,120000],
    "Experience": [1,2,4,6,8,10],
    "Department": [
        "IT",
        "IT",
        "HR",
        "HR",
        "Finance",
        "Finance"
    ]
}

df = pd.DataFrame(data)

sns.pairplot(
  df,
  hue="Department"
)

plt.show()
````

## Practical 3: Pair Plot with Different Diagonal Plot

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Age": [22,25,28,30,35,40],
    "Experience": [1,2,4,6,8,10],
    "Salary": [25000,35000,50000,70000,90000,120000]
}

df = pd.DataFrame(data)

sns.pairplot(
    df,
    diag_kind="kde"
)

plt.show()
````
