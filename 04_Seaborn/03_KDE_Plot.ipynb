# KDE Plot (Kernel Density Estimation)

## Introduction

A KDE (Kernel Density Estimation) Plot is a statistical visualization used to estimate the probability density distribution of a continuous variable.

It is similar to a Histogram, but instead of displaying bars, it displays a smooth continuous curve that represents the distribution of the data.

KDE Plots help Data Scientists understand how data is distributed and where observations are concentrated.

---

# Why KDE Plot is Important?

When analyzing data, understanding the distribution of features is very important.

KDE Plots help us:

* Understand data distribution
* Identify data concentration
* Detect skewness
* Compare distributions
* Analyze continuous variables

Unlike Histograms, KDE Plots do not depend on bin sizes and provide a smoother representation of the data.

---

# Histogram vs KDE Plot

| Histogram       | KDE Plot          |
| --------------- | ----------------- |
| Uses Bars       | Uses Smooth Curve |
| Depends on Bins | No Bins Required  |
| Shows Frequency | Shows Density     |
| Less Smooth     | More Smooth       |

---

# Understanding KDE Plot

A KDE Plot estimates how likely values are to occur within a dataset.

The graph consists of:

```text
X-axis → Data Values

Y-axis → Density
```

Higher peaks indicate areas where many observations are concentrated.

Lower peaks indicate fewer observations.

---

# What Does Density Mean?

Density represents the concentration of data points.

A higher density means:

```text
More observations in that region
```

A lower density means:

```text
Fewer observations in that region
```

---

# Why Use KDE Instead of Histogram?

Histograms can look different depending on the number of bins selected.

KDE Plots eliminate this problem by creating a smooth estimate of the distribution.

This makes it easier to:

* Understand trends
* Compare datasets
* Analyze feature distributions

---



## Practical 1: Basic KDE Plot

````python
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 12, 15, 18, 20,
        22, 25, 28, 30, 35]

sns.kdeplot(data)

plt.title("Basic KDE Plot")

plt.show()
````

## Practical 2: KDE Plot with Fill

````python
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 12, 15, 18, 20,
        22, 25, 28, 30, 35]

sns.kdeplot(
    data,
    fill=True
)

plt.title("KDE Plot with Fill")

plt.show()
````

## Practical 3: KDE Plot Using NumPy

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = np.random.normal(
    50,
    10,
    1000
)

sns.kdeplot(
    data,
    fill=True
)

plt.title("Normal Distribution KDE")

plt.show()
````

## Practical 4: Real Data Science Example

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

employee_data = {
    "Salary": [
        25000,30000,35000,40000,
        45000,50000,55000,60000,
        65000,70000,75000,80000,
        85000,90000,95000,100000
    ]
}

df = pd.DataFrame(employee_data)


sns.kdeplot(
  df["Salary"],
  fill=True
)

plt.title("Salary Distribution")

plt.xlabel("Salary")

plt.show()
````

# Applications in Data Science

KDE Plots are commonly used for:

### Exploratory Data Analysis (EDA)

Understanding feature distributions before modeling.

### Feature Analysis

Checking whether features are normally distributed.

### Data Preprocessing

Identifying skewed features that may require transformation.

### Outlier Detection

Finding unusual observations in the dataset.

---

# Applications in Machine Learning

Before training Machine Learning models, KDE Plots help analyze:

* Feature distributions
* Data quality
* Continuous variables
* Data imbalance

Understanding the distribution of features often improves model performance.

---

# Advantages of KDE Plot

* Smooth visualization
* Easy to interpret
* Better distribution analysis
* No dependence on bins
* Useful for large datasets

---

# Limitations of KDE Plot

* Not suitable for categorical data
* Requires numerical data
* Less effective for very small datasets
* May hide individual observations
