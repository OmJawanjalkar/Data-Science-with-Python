# Violin Plot

## Introduction

A Violin Plot is a statistical visualization that combines the features of a Box Plot and a KDE Plot.

It shows:

* Distribution of data
* Density of data
* Median value
* Spread of data

The shape of the violin represents the density of the data at different values.

---

# Why Violin Plot is Important?

A Box Plot only shows summary statistics such as:

* Median
* Quartiles
* Outliers

A KDE Plot only shows the density distribution.

A Violin Plot combines both and provides more information about the data distribution.

---

# Box Plot vs KDE Plot vs Violin Plot

| Plot        | Information Provided        |
| ----------- | --------------------------- |
| Box Plot    | Median, Quartiles, Outliers |
| KDE Plot    | Density Distribution        |
| Violin Plot | Box Plot + KDE Plot         |

---

# Understanding Violin Plot

A Violin Plot displays:

```text
Center Line → Median

Width of Shape → Density

Wider Area → More Data Points

Narrow Area → Fewer Data Points
```

The wider the violin, the higher the concentration of data values.

---

# Why Use Violin Plot?

Violin Plots help answer:

* How is the data distributed?
* Where are most observations concentrated?
* Is the data symmetric?
* Are there multiple peaks?
* How do groups compare?

---

# Applications in Data Science

Violin Plots are commonly used for:

### Exploratory Data Analysis (EDA)

Understanding feature distributions.

### Feature Analysis

Comparing distributions across categories.

### Data Preprocessing

Detecting unusual distributions.

### Statistical Analysis

Analyzing data spread and density.

---

# Applications in Machine Learning

Before model training, Violin Plots help:

* Analyze feature distributions
* Compare target classes
* Detect skewness
* Understand data variability

---

# Violin Plot Syntax

```python
sns.violinplot(data=data)
```

---

# Common Parameters

| Parameter | Purpose                  |
| --------- | ------------------------ |
| data      | Input Data               |
| x         | Category Variable        |
| y         | Numerical Variable       |
| inner     | Show Internal Statistics |
| palette   | Change Colors            |

---



## Practical 1: Basic Violin Plot

````python
import seaborn as sns
import matplotlib.pyplot as plt

data = [10,15,20,25,30,35,40,45,50]

sns.violinplot(data=data)

plt.title("Basic Violin  Plot")
plt.show()
````

## Practical 2: Violin Plot with Inner Statistics

````python
import seaborn as sns
import matplotlib.pyplot as plt

data = [10,15,20,25,30,35,40,45,50]

sns.violinplot(
    data=data,
    inner="quartile"
)

plt.title("Violin Plot with Quartiles")

plt.show()
````

## Practical 3: Category-wise Violin Plot

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Department": [
        "IT","IT","IT",
        "HR","HR","HR",
        "Finance","Finance","Finance"
    ],
    "Salary": [
        50000,60000,70000,
        40000,45000,50000,
        55000,65000,75000
    ]
}

df = pd.DataFrame(data)

sns.violinplot(
    x="Department",
    y="Salary",
    data=df,
    inner="quartile"
)

plt.title("Salary Distribution by Department")

plt.show()
````
